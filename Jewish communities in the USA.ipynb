### read the table into dataframe  
in the ROW loop, we are going over all the rows in the html talbe and put it into array named data
in the COLUMN loop, we are going over the columns: City, Country, Number. 
Now, we have to clean, all the cities that are not in the US.
also, we need to clean the text:
1. clean then refernces number like [1]
2. convert the Number column to Int

```python
data= [None, None, None]
num_of_jew_df = pd.DataFrame(columns=['City', 'State',  'Number', 'Latitude', 'Longtitude'], index=np.arange(0, 54)) 
row_marker = 0
for row in table.find_all('tr'):     
    column_marker = 0    
    columns = row.find_all('td')          
    for column in columns:                
        #print (column.get_text(), row.get('href'))          
        data[column_marker]=column.get_text().replace( '\n', '')
        column_marker+=1            
    if data[0] !=None:
       if 'United States' in data[1]: 
          city = data[0].replace(data[0][data[0].find('['):data[0].find(']')+1], '')          
          while  city.find('[')>0:
                 city = city.replace(city[city.find('['):city.find(']')+1], '')
          num_of_jew_df.iat[row_marker, 0]= city
          number_of_pop = data[2].replace(data[2][data[2].find('['):data[2].find(']')+1], '')          
          while number_of_pop.find(',')>0:                
                number_of_pop = number_of_pop.replace(',','')                
          num_of_jew_df.iat[row_marker, 2]= int(number_of_pop)
          row_marker +=1
    
    
num_of_jew_df
```

```python
# one hot encoding
community_onehot = pd.get_dummies(community_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
community_onehot['City'] = community_venues['City'] 

# move neighborhood column to the first column
fixed_columns = [community_onehot.columns[-1]] + list(community_onehot.columns[:-1])
community_onehot = community_onehot[fixed_columns]

community_onehot
```

```python
num_top_venues = 5

for hood in community_grouped['City']:
    print("----"+hood+"----")
    temp = community_grouped[community_grouped['City'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

### Let's put that into a pandas dataframe  
##### First, let's write a function to sort the venues in descending order.

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 5

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['City']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
community_neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
community_neighborhoods_venues_sorted['City'] = community_grouped['City']

for ind in np.arange(community_grouped.shape[0]):
    community_neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(community_grouped.iloc[ind, :], num_top_venues)

community_neighborhoods_venues_sorted
```

# Cluster Neighborhoods

### Run k-means to cluster the neighborhood into 5 clusters

```python
# set number of clusters
kclusters = 4

community_grouped_clustering = community_grouped.drop('City', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(community_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10]
```

```python
# add clustering labels
community_neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

community_merged = num_of_jew_df

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
community_merged = community_merged.join(community_neighborhoods_venues_sorted.set_index('City'), on='City')

community_merged
```

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=4)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
community_merged['Cluster Labels'].dropna(inplace=True)
for lat, lon, poi, cluster in zip(community_merged['Latitude'], community_merged['Longtitude'], community_merged['City'], community_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[int(cluster)-1],
        fill=True,
        fill_color=rainbow[int(cluster)-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
